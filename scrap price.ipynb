Data preprocesing

_for Object data

```python
df.drop(['ID', 'name'], axis=1, inplace=True)
```

```python
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
```

```python
# 2. Value Counts Analysis
print("\nValue Counts for Categorical Columns:")
for col in categorical_cols:
    print(f"\n---------")
    print(df[col].value_counts())
    print(f"Unique values: {df[col].nunique()}")
```

```python
plt.figure(figsize=(15, 10))
for i, col in enumerate(categorical_cols, 1):
    plt.subplot(3, 3, i)
    sns.countplot(data=df, x=col)
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

_feature eng for identified object data

```python
df['doornumbers'] = df['doornumbers'].map({'two': 2, 'four': 4})
```

```python
df.drop(['enginelocation'], axis=1, inplace=True)
```

```python
df['carbody'] = df['carbody'].apply(lambda x: x if x in ['sedan', 'hatchback'] else 'other')
df['enginetype'] = df['enginetype'].apply(lambda x: 'ohc' if x == 'ohc' else 'other')
df['cylindernumber'] = df['cylindernumber'].apply(lambda x: 'four' if x == 'four' else 'other')
df['fuelsystem'] = df['fuelsystem'].apply(lambda x: x if x in ['mpfi', '2bbl'] else 'other')
```

```python
df = pd.get_dummies(df, columns=['fueltypes','aspiration','drivewheels',
                                'carbody', 'enginetype','cylindernumber','fuelsystem'], dtype=int)
```

EDA

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], kde=True)
plt.title('Price Distribution')
plt.show()
```

```python
plt.figure(figsize=(24, 16)) 
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
correlation = df.corr(numeric_only=True)[['price']].sort_values('price', ascending=False)

plt.figure(figsize=(10, 8))
sns.barplot(
    x='price', 
    y=correlation.index, 
    data=correlation, 
    hue=correlation.index,  
    palette='viridis', 
    legend=False  
)
plt.title('Correlation with Price')
plt.xlabel('Correlation Coefficient')
plt.ylabel('Features')
plt.show()
```
